```python
import json
import numpy as np
from tqdm import tqdm
```

```python
msrvtt_annotations_path = r'/ssd_scratch/cvit/varun/MSRVTT/annotation/MSR_VTT.json'

msrvtt_annotations = json.load(open(msrvtt_annotations_path, 'r'))
```

```python
annots = msrvtt_annotations['annotations']
```

```python
new_annots = {}
for annotation in tqdm(annots):
    if annotation['image_id'] not in new_annots:
        new_annots[annotation['image_id']] = []
    new_annots[annotation['image_id']].append({key: value for key, value in annotation.items() if key != 'image_id'})
```
